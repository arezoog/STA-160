## Merge all **cleaned** files into a master df

```python
import pandas as pd
import glob
import os

clean_dir = r"/Users/arezoog/Desktop/STA 160/Clean Data"

clean_files = glob.glob(os.path.join(clean_dir, "*.csv"))
clean_dfs = [pd.read_csv(f) for f in clean_files]

master_clean_df = pd.concat(clean_dfs, ignore_index=True)
print("Combined Cleaned Rows:", len(master_clean_df))

master_clean_path = os.path.join(clean_dir, "EQR_master_clean.csv")
master_clean_df.to_csv(master_clean_path, index=False)

print("Saved master clean file to:", master_clean_path)
```

## Merge all **output** files into a master df

```python
output_dir = r"/Users/arezoog/Desktop/STA 160/Output DFs"

output_files = glob.glob(os.path.join(output_dir, "*.csv"))
output_dfs = [pd.read_csv(f) for f in output_files]

master_output_df = pd.concat(output_dfs, ignore_index=True)
print("Combined Raw Output Rows:", len(master_output_df))

master_output_path = os.path.join(output_dir, "EQR_master_output.csv")
master_output_df.to_csv(master_output_path, index=False)

print("Saved master output file to:", master_output_path)
```
